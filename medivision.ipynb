```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from langchain_community.llms import HuggingFacePipeline

model_name = "google/flan-t5-large"

# Load tokenizer & model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Create Hugging Face generation pipeline
summarizer_pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=512,
)

# Wrap in LangChain
llm = HuggingFacePipeline(pipeline=summarizer_pipeline)
```

```python
prompt = PromptTemplate(
    template="""
You are a medical expert. Convert this medical report into very simple and
human-understandable explanation. Avoid medical jargon.

Report:
{text}

Explanation:
""",
    input_variables=["text"]
)

output_parser = StrOutputParser()

chain = prompt | llm | output_parser
```

```python
import pytesseract
from PIL import Image
import numpy as np

def extract_text(image):
    img = Image.fromarray(image)
    text = pytesseract.image_to_string(img)
    return text
```

```python
def process_medical_report(image):
    extracted_text = extract_text(image)

    if not extracted_text.strip():
        return "❌ Could not extract any text. Try a clearer image."

    summary = chain.invoke({"text": extracted_text})

    return f"### 📝 Extracted Text:\n{extracted_text}\n\n### 📌 Summary:\n{summary}"
```

```python
import gradio as gr

ui = gr.Interface(
    fn=process_medical_report,
    inputs=gr.Image(type="numpy", label="Upload Medical Report Image"),
    outputs="markdown",
    title="Medical Report Interpreter",
    description="Upload a photo of a medical report. AI will extract the text using OCR and explain it in simple language."
)

ui.launch()
```
